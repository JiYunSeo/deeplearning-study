```python
A=np.array(data)
```

```python
A
```

```python
import scipy.interpolate
```

```python
from scipy.interpolate import interp2d
f = interp2d(np.arange(0,500,100), np.arange(0,600,100), A)
aa = f(np.arange(500), np.arange(600))
```

```python
plt.subplots(figsize=(10,8))
heatmap = sns.heatmap(data, annot=True, fmt='.0f', cmap='jet', annot_kws={"size": 15}, cbar=True, vmin=0, vmax=100)
plt.xlabel('back', fontsize=20)
plt.ylabel('left', fontsize=20)
plt.axis('off')
plt.show()

plt.subplots(figsize=(10,8))
colormap = sns.heatmap(aa, cmap='jet',cbar=True, vmin=0, vmax=100)
plt.xlabel('back', fontsize=20)
plt.ylabel('left', fontsize=20)
plt.axis('off')
plt.show()

#https://pod.hatenablog.com/entry/2018/09/20/212527
```
